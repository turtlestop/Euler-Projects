```python
# If the numbers 1 to 5 are written out in words: 
#     one, two, three, four, five,
#     then there are 3 + 3 + 5 + 4 + 4 = 19 letters used 
#     in total.

# If all the numbers from 1 to 1000 (one thousand) 
# inclusive were written out in words, how many letters
# would be used?
```

```python
units = ['one','two','three','four','five','six','seven','eight','nine']
unit_count = sum([len(word) for word in units])
tens = ['twenty', 'thirty','forty','fifty','sixty', 'seventy', 'eighty','ninety']
teens = ['eleven','twelve','thirteen','fourteen','fifteen', 'sixteen','seventeen','eighteen','nineteen']
teen_count = sum([len(word) for word in teens])
def count():
    total = 0
    hundred = 0
#     numbers up to 19
    hundred += unit_count +1
    hundred += teen_count
#     adding 20 - 99:
    for tenword in ten:
        hundred += 10*len(tenword) + unit_count
    for word in units:
        total += len(word) + 9*len('hundred and') + hundred
    total += 9*len('hundred') + units_count
    total += len('one thousand')
    return total
    
```

```python
def wordy_number(n):
    name = ""
    if n == 1000:
        return "one thousand"
    if n>100:
        hundred_digit = units[n/100-1]
        name += hundred_digit + " hundred"
    n = n%100
    if n == 0:
        return name
    else:
        name += ' and '
    if n<10:
        name += units[n-1]
    elif n>10 and n<20:
        name += teens[n-11]
    if n>20:
        unit_digit = n%10
        if unit_digit == 0:
            name += tens[(n/10)-2]
        else:
            name += tens[(n/10)-2] + ' ' + units[unit_digit -1]
    return name
```
